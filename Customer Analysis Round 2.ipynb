```python
print(df['Vehicle Type'].unique())
```

```python
df['Vehicle Type']= df['Vehicle Type'].fillna('not A')
#I decided to put 'not A' because apperently it is not A but something else
```

```python
print(df['State'].unique())
print(df['State'].value_counts())
df['State'] = df['State'].fillna('California')
print(df['State'].value_counts())
#I decided to put 'California' in the nan values because it is the most frequent value
```

```python
print(df['Response'].unique())
print(df['Response'].value_counts())
df['Response'] = df['Response'].fillna('No') 
#I decided to put 'No' in the nan values because it is the most frequent value
```

```python
print(df['Number of Open Complaints'].unique())
print(df['Number of Open Complaints'].value_counts())
#df[(df['Months Since Last Claim'].isna()==True) & (df['Number of Open Complaints'].isna()=True)]
df[df['Months Since Last Claim'].isna()==True]
```

It seems that every NaN in 'Months Since Last Claim' belongs to a NaN in 'Number of Open Complaints', so i decide to put both to 0.


```python
df['Months Since Last Claim'] = df['Months Since Last Claim'].fillna(0)
df['Number of Open Complaints'] = df['Number of Open Complaints'].fillna(0)
```
